```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/vehicle_emissions.csv')
```

```python
df.dropna(inplace=True)
value = 0
df.fillna(value, inplace=True)
```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python
#Histogram of CO2 Emissions by Fuel Type
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="co2_emissions", hue="fuel_type", element="step")
plt.title("Histogram of CO2 Emissions by Fuel Type")
plt.xlabel("CO2 Emissions (g/km)")
plt.ylabel("Count")
plt.show()
```

```python
#Scatter Plot: Engine Size vs CO2 Emissions by Fuel Type
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="engine_size", y="co2_emissions", hue="fuel_type")
plt.title("Engine Size vs CO2 Emissions by Fuel Type")
plt.xlabel("Engine Size (L)")
plt.ylabel("CO2 Emissions (g/km)")
plt.show()
```

```python
#Box Plot: CO2 Emissions by Emission Category
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="emission_category", y="co2_emissions")
plt.title("CO2 Emissions by Emission Category")
plt.xlabel("Emission Category")
plt.ylabel("CO2 Emissions (g/km)")
plt.show()
```

```python
#Bar Chart: Average CO2 Emissions by Fuel Type
avg_emissions = df.groupby("fuel_type")["co2_emissions"].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(data=avg_emissions, x="fuel_type", y="co2_emissions")
plt.title("Average CO2 Emissions by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average CO2 Emissions (g/km)")
plt.show()
```

```python
# Create a violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x="fuel_type", y="co2_emissions")

plt.title("Violin Plot of CO2 Emissions by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("CO2 Emissions (g/km)")
plt.show()
```
